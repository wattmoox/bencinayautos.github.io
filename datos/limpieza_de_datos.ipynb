# Codigo de limpieza de los datos 

Creado con el objetivo de limpiar las variables, calcular los datos a ocupar y dejar en el tipo correcto cada variable

```python
#librerias a utilizar
import pandas as pd

#importacion del archivo autos.xlsx
df = pd.read_excel('bdd.xlsx')
df.head()
```

```python
#limpieza de tipos de datos
df.info()
```

Como podemos visualizar la variables numeras como `consumo mixto aproximado` y `precio aproximado litro gasolina` estan en variables object. Oara poder hacer el calculo de la variable `litros cargados con 50 dolares ` necesitaremos pasarlas a int limpiando cada digito de su agregado

```python
#quitamos la frase " km/L" de los datos del atributo "Consumo Mixto Aprox."
df['Consumo Mixto Aprox.'] = df['Consumo Mixto Aprox.'].str.replace(' km/L', '')

#convertimos la variable "Consumo Mixto Aprox." a tipo numérico
df['Consumo Mixto Aprox.'] = pd.to_numeric(df['Consumo Mixto Aprox.'], errors='coerce')

#verificamos los cambios realizados
df.info()
```

```python
#mismo procedimiento con Precio Aprox. Litro Gasolina (USD)
#eliminamos letras o elementos $ que no sean numeros en la variable a convertir
df['Precio Aprox. Litro Gasolina (USD)'] = df['Precio Aprox. Litro Gasolina (USD)'].str.extract(r'(\d+\.?\d*)')

#convertir la variable a numerica
df['Precio Aprox. Litro Gasolina (USD)'] = pd.to_numeric(df['Precio Aprox. Litro Gasolina (USD)'], errors='coerce')

#ver como quedaron las modificaciones
df.head()

```

```python
#hacemos el calculo a agregar en la variable Litros cargados con 50 dolares (dolares/precio)
df['Litros cargados con 50 dolares'] = 50 / df['Precio Aprox. Litro Gasolina (USD)']
#hacemmos el calculo a agregar en la variable Kilometros recorridos con 50 dolares (litros cargados/consumo) 
df['Kilometros cargados con 50 dolares'] = df['Litros cargados con 50 dolares'] * df['Consumo Mixto Aprox.']
```

Ya con nuestra base de datos totalmente limpia y utilizable solo nos queda filtrar que datos son escenciales para el desarrollo de nuestra pagina

```python
#usamos solo País	Sedán Más Vendido (Combustión)	Consumo Mixto Aprox.	Precio Aprox. Litro Gasolina (USD)	Litros cargados con 50 dolares	Kilometros cargados con 50 dolares
df_final = df[['País', 'Sedán Más Vendido (Combustión)', 'Consumo Mixto Aprox.', 'Precio Aprox. Litro Gasolina (USD)', 'Litros cargados con 50 dolares', 'Kilometros cargados con 50 dolares']]
```

```python
# importamos las modificaciones hechas a un nuevo archivo 
df_final.to_json("datos_limpios.json", orient="records", force_ascii=False, indent=2)
```
