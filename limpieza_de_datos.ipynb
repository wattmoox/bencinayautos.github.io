```python
#librerias a utilizar
import pandas as pd

#importacion del archivo autos.xlsx
df = pd.read_excel('bdd.xlsx')
df.head()
```

```python
#limpieza de tipos de datos
df.info()
```

Como podemos visualizar la variables numeras como `consumo mixto aproximado` y `precio aproximado litro gasolina` estan en variables object. Oara poder hacer el calculo de la variable `litros cargados con 50 dolares ` necesitaremos pasarlas a int limpiando cada digito de su agregado

```python
#quitamos la frase " km/L" de los datos del atributo "Consumo Mixto Aprox."
df['Consumo Mixto Aprox.'] = df['Consumo Mixto Aprox.'].str.replace(' km/L', '')

#convertimos la variable "Consumo Mixto Aprox." a tipo numérico
df['Consumo Mixto Aprox.'] = pd.to_numeric(df['Consumo Mixto Aprox.'], errors='coerce')

#verificamos los cambios realizados
df.info()
```

```python
#mismo procedimiento con Precio Aprox. Litro Gasolina (USD)
#eliminamos letras o elementos $ que no sean numeros en la variable a convertir
df['Precio Aprox. Litro Gasolina (USD)'] = df['Precio Aprox. Litro Gasolina (USD)'].str.replace('$', '', '(Subsidio)', '(Oficial)', '(Intl.) / $0.04 (Sub)')

#convertir la variable a numerica
df['Precio Aprox. Litro Gasolina (USD)'] = pd.to_numeric(df['Precio Aprox. Litro Gasolina (USD)'], errors='coerce')

#ver como quedaron las modificaciones
df.

```
